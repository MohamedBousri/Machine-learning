```python
import pandas as pd
import numpy as np 

header=["Class Name:(democrat, republican)"
   ,"handicapped-infants"
   ," water-project-cost-sharing"
   ," adoption-of-the-budget-resolution"
   ," physician-fee-freeze"
   ," el-salvador-aid"
   ," religious-groups-in-schools"
   , "anti-satellite-test-ban"
   , "aid-to-nicaraguan-contras"
  , "mx-missile"
  ," immigration"
  , "synfuels-corporation-cutback"
  , "education-spending"
  , "superfund-right-to-sue"
  , "crime"
  , "duty-free-exports"
  , "export-administration-act-south-africa"
]
data=pd.read_csv("mltpdataset/house-votes-84.data",names=header); 
data.head()
```

```python
for n in header : 
    data=data[data[n]!="?"]
data.head()
```

```python
for n in header : 
    data=data[data[n]!="?"]

data=data.replace('y',1)
data=data.replace('n',0)
data.head()
```

```python
X=data.values[:,1:16]
Y=data.values[:,0]
```

Logistique regression 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.2, random_state = 100)

model = LogisticRegression()
model.fit(X_train, y_train)

predicted_classes = model.predict(X_test)
h=model.predict(X_train)


print("Train Accuracy is ", accuracy_score(y_train, h)*100)
print("Test Accuracy is ", accuracy_score(y_test,predicted_classes)*100)
```

# Filter feature selection


SelectKbest

```python
import pandas as pd  
import numpy as np  
from sklearn.model_selection import train_test_split  
from sklearn.feature_selection import SelectKBest, f_classif,chi2
from collections import Counter

model = LogisticRegression()

#score_function=f_classif ANOVA F-VALUE


selector=SelectKBest(score_func=f_classif,k=8)
X_Selectk1=selector.fit_transform(X,Y)
X_train, X_test, y_train, y_test = train_test_split(X_Selectk1, Y, test_size=0.2, random_state=100) 

model.fit(X_train,y_train) 

predicted_classes = model.predict(X_test)
h = model.predict(X_train)
print("Train Accuracy after feature selection with SelectKbest (f_classif) is ", accuracy_score(y_train,h)*100)
print("Test Accuracy after feature selection with SelectKbest (f_classif) is ", accuracy_score(y_test,predicted_classes)*100)

#score_function=CHI2

selector=SelectKBest(score_func=chi2,k=8)
X_Selectk2=selector.fit_transform(X,Y)
X_train, X_test, y_train, y_test = train_test_split(X_Selectk2, Y, test_size=0.2, random_state=100) 

model.fit(X_train,y_train) 

predicted_classes = model.predict(X_test)
h = model.predict(X_train)
print("Train Accuracy after feature selection with SelectKbest (chi2) is ", accuracy_score(y_train,h)*100)
print("Test Accuracy after feature selection with SelectKbest (chi2) is ", accuracy_score(y_test,predicted_classes)*100)

```

# Wrapper feature selection

 RFE

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import RFE
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt


#model2 = DecisionTreeClassifier(criterion ="entropy", max_depth=10 , splitter="best" , min_impurity_split=0.2 )
listMin =[]
listAcc =[]
model2 = LogisticRegression()

for i in range (1,7) :
    rfe = RFE(model2, i*2)
    X_rfe = rfe.fit_transform(X,Y)
    X_train, X_test, y_train, y_test = train_test_split(X_rfe, Y, test_size=0.22, random_state=100) 
    model2.fit(X_train,y_train) 
    predicted_classes = model2.predict(X_test)
    listMin.append(i*2)
    listAcc.append(accuracy_score(y_test,predicted_classes)*100)

```

```python
listAcc = np.asarray(listAcc)
listMin = np.asarray(listMin)

listAcc = listAcc[:,np.newaxis]
listMin = listMin[:,np.newaxis]
plt.title("changment du Test accurancy en fonction du nombre minimum d'attributs donné a RFE  \n ",fontsize=10)
plt.scatter(listMin,listAcc)
```
